### Figure 1

```python

%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

# enter additional commands here to reproduce the figure below
plt.figure()
plt.xlabel("Time")
plt.ylabel("Value")
plt.plot(a, marker='.', markersize=10, color='black')
plt.show()
# NOTE!
# - Entering commands and running the cell may cause the figure to disapper
# - If the figure disappears, use Assn-3.html as a reference
```

### Figure 2

Use `data-fig2.npy` for this figure.

```python
a   = np.load( 'data-fig2.npy' )
print(a.shape)


plt.xlabel("X value")
plt.ylabel("Y value")
plt.plot(a[:, 0], label=f"blue", color="Blue")
plt.plot(a[:, 1], label=f"red", color="Red")
plt.plot(a[:, 2], label=f"green", color="Green")
plt.legend(loc="upper left")
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
a=np.load( 'data-fig3.npy' )
print(a.shape)

x=a[:, 0]
y=a[:, 1]
plt.figure()
plt.xlabel("X value")
plt.ylabel("Y value")
plt.scatter(x, y, s=200, c='blue', alpha=0.3)
plt.show()
```

### Figure 4

Use `data-fig4.npy` for this figure.

```python
a=np.load( 'data-fig4.npy' )
x = np.arange(len(a))

plt.figure()
plt.xlabel("Category")
plt.ylabel("Value")
plt.bar(x, a, color='gray')
plt.show()
```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
a=np.load( 'data-fig5.npy' )
print(a.shape)

plt.figure()
plt.pcolor(a, cmap='RdBu')
plt.colorbar()
plt.show()
```
